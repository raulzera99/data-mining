```python
import pandas as pd 
```

```python
businessAnalyst = pd.read_csv('csv/BusinessAnalyst.csv')
dataAnalyst = pd.read_csv('csv/DataAnalyst.csv')
dataEngineer = pd.read_csv('csv/DataEngineer.csv')
dataScientist = pd.read_csv('csv/DataScientist.csv')
```

```python
# excluir colunas que não são comuns a todos os dataframes
businessAnalyst = businessAnalyst.drop(columns=['Unnamed: 0', 'index'])
dataAnalyst = dataAnalyst.drop(columns=['Unnamed: 0'])
# dataEngineer = dataEngineer.drop(columns=['Unnamed: 0', 'index'])

```

```python
df = [businessAnalyst, dataAnalyst, dataEngineer]

df = pd.concat([businessAnalyst, dataAnalyst, dataEngineer], ignore_index=True)

df.to_csv('csv/combined.csv', index=False)
```

```python
# transformar os tipos de dados
df['Job Title'] = df['Job Title'].astype('str')
df['Salary Estimate'] = df['Salary Estimate'].astype('str')
df['Job Description'] = df['Job Description'].astype('str')
df['Rating'] = df['Rating'].astype('str')
df['Company Name'] = df['Company Name'].astype('str')
df['Location'] = df['Location'].astype('str')
df['Headquarters'] = df['Headquarters'].astype('str')
df['Size'] = df['Size'].astype('str')
df['Founded'] = df['Founded'].astype('str')
df['Type of ownership'] = df['Type of ownership'].astype('str')
df['Industry'] = df['Industry'].astype('str')
df['Sector'] = df['Sector'].astype('str')
df['Revenue'] = df['Revenue'].astype('str')
df['Competitors'] = df['Competitors'].astype('str')
df['Easy Apply'] = df['Easy Apply'].astype('str')

df.head()
```

```python
# tratar coluna Salary Estimate excluindo (Glassdoor est.) 
df['Salary Estimate'] = df['Salary Estimate'].str.replace('Glassdoor est.', '')
# adicionar coluna salario minimo e salario maximo e preencher com valores da coluna Salary Estimate
df[['Salary Min', 'Salary Max']] = df['Salary Estimate'].str.split('-', expand=True)

df.head()
```
